```python
### 
# x = beta_rec - beta_true 
from scipy import integrate
parnames = ['mean', 'sigma', 'sigma_ratio', 'asy_factor', 'fraccore']



def prob_resolution(x, mean, sigma, sigma_ratio, asy_factor, fraccore):                                                                                                                                                                      
    coregaus = gaussian(x, mean, sigma)                                                                                                                                                                                                           
    asygaus = asy_gaussian_1d(x, mean,  sigma_ratio * sigma, asy_factor)                                                                                                                                                                          
    pdf = (fraccore * coregaus + (1 - fraccore) * asygaus)                                                                                                                                                                                 
    return pdf 
    
df_probpars = np.load('/home/manbing/Documents/lithiumanalysis/scripts/plots/unfold/beta/AglBe7_polypar.npz')

mass_nucleon_gev = 0.9314941 #in GeV 

def dEdbeta(beta):
    return mass_nucleon_gev * beta * (1-beta**2)**(-3/2)

def rate_func(b_true, b_rec, phi, c, rigidity_scale, gamma):
    flux = power_law(calc_ekin_from_beta(b_true), phi, c, rigidity_scale, gamma) * dEdbeta(b_true)
    #fluxTimeAcc = isoflux_powerlaw['Be7'](calc_ekin_from_beta(b_true)) * spline_subgraph_acc['Agl']['Be7'](np.log(calc_ekin_from_beta(b_true))) 
    fluxTimeAcc = flux * spline_subgraph_acc['Agl']['Be7'](np.log(calc_ekin_from_beta(b_true))) 
    prob_par = dict()
    for par in parnames:
        prob_par[par] = np.poly1d(df_probpars[par])(np.log(b_true))
    event_rate = fluxTimeAcc * prob_resolution(b_true - b_rec, prob_par['mean'], prob_par['sigma'], prob_par['sigma_ratio'],prob_par['fraccore'],prob_par['asy_factor'] )
    return event_rate
    
beta_range = [0.97, 0.9999]
xbeta_binning = Binning(calc_beta_from_ekin(fbinning_energy_Agl()))
   
print(isoflux_param_dict['Be7'])
flux_pars =  list(isoflux_param_dict['Be7'].values())
#print(xbeta_binning.bin_centers[binrange[0]: binrange[1]])
#print(calc_ekin_from_beta(xbeta_binning.bin_centers[binrange[0]: binrange[1]]))


    
```

```python
def get_predict_rate(xbeta_binning, beta_range, flux_pars):
    binrange = xbeta_binning.get_indices(beta_range) 
    xvalue_beta = xbeta_binning.bin_centers[binrange[0]: binrange[1]] 
    predict_rate = np.zeros_like(xvalue_beta)
    predict_rate_err = np.zeros_like(xvalue_beta)
    #print(xbeta_binning.edges)
    #print(xvalue_beta)
    for ip, ibin in enumerate(range(binrange[0], binrange[1])): 
        #print(ibin)
        #print('min:', min(ibin+3, len(xbeta_binning.edges)-2))
        predict_rate[ip], predict_rate_err[ip] = integrate.nquad(rate_func, [[xbeta_binning.edges[ibin-2], xbeta_binning.edges[min(ibin+3, len(xbeta_binning.edges)-2)]],[xbeta_binning.edges[ibin],xbeta_binning.edges[ibin+1]]], args=(flux_pars), opts=dict(epsrel=1e-3))
    return predict_rate

#predict_rate = get_predict_rate(xbeta_binning, beta_range, flux_pars)
#print(predict_rate)
graph_predict_rate = MGraph()
```
