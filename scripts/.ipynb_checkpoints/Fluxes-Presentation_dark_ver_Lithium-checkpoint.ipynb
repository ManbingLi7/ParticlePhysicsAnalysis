### Apply FFA modulation

```python
# Phi=679.e-3
Phi=phi_5yr/1000
# Phi=phi_8hyr/1000

ek_galprop_mod_ek = [ek_galprop.copy() for i in range(2)]
ek_galprop_mod_rig = [ek_galprop.copy() for i in range(2)]
fluxes_galprop_mod_rig = [fluxes_galprop[0].copy(), fluxes_galprop[1].copy()]

z = 3
mp=0.93827208816 #proton mass in GeV/c2 PDG 2022
mn=0.93956542052 #neutron mass in GeV/c2 PDG 2022

for ia, a in enumerate((6, 7)):
    m=mp*z+mn*(a-z) #should substract binding energy about 8MeV/A for Z>20 and less below see http://ne.phys.kyushu-u.ac.jp/seminar/MicroWorld3_E/3Part2_E/3P23_E/nuclear_mass_E.htm
    benergy = (31.994564e-3, 39.244526e-3)[ia] # binding energy in GeV/c^2  from  http://barwinski.net
    
    m=m-benergy

    m_nucl=m/a #average mass per nucleon

    for i in range(len(ek_galprop)):
        mod_ekn = ek_galprop[i] - z*Phi/a #Ek/n in GeV/n
        if mod_ekn >= 0:
            mod_flux = fluxes_galprop[ia][i]*(mod_ekn*(mod_ekn+2*m_nucl)) / (ek_galprop[i] * (ek_galprop[i] + 2*m_nucl))
            
            ek_galprop_mod_ek[ia][i]  = mod_ekn
            ek_galprop_mod_rig[ia][i] = wphysics.CalRigwithEk(Ek=mod_ekn, A=a, C=z)
            
            dek = (ek_galprop_mod_ek[ia][i]-ek_galprop_mod_ek[ia][i-1])
            drig = ek_galprop_mod_rig[ia][i] - ek_galprop_mod_rig[ia][i-1]
            fluxes_galprop_mod_rig[ia][i] = mod_flux*dek/drig
```

```python
istart = 65
fluxes_galprop_mod_rig_tck = [interpolate.UnivariateSpline(np.log(ek_galprop_mod_rig[ia][istart:]), np.log(fluxes_galprop_mod_rig[ia][istart:]), k=3, s=0.001, ext=3) for ia in range(2)]
```

```python
# Phi=671.e-3
# Phi=phi_5yr/1000
Phi=phi_8hyr/1000

ek_galprop_mod = [ek_galprop.copy() for i in range(2)]
fluxes_galprop_mod = [fluxes_galprop[0].copy(), fluxes_galprop[1].copy()]

z = 3
mp=0.93827208816 #proton mass in GeV/c2 PDG 2022
mn=0.93956542052 #neutron mass in GeV/c2 PDG 2022

for ia, a in enumerate((6, 7)):
    m=mp*z+mn*(a-z) #should substract binding energy about 8MeV/A for Z>20 and less below see http://ne.phys.kyushu-u.ac.jp/seminar/MicroWorld3_E/3Part2_E/3P23_E/nuclear_mass_E.htm
    benergy = (31.994564e-3, 39.244526e-3)[ia] # binding energy in GeV/c^2  from  http://barwinski.net
    
    m=m-benergy

    m_nucl=m/a #average mass per nucleon

    for i in range(len(ek_galprop)):
        mod_ekn = ek_galprop[i] - z*Phi/a #Ek/n in GeV/n
        if mod_ekn >= 0:
            mod_flux = fluxes_galprop[ia][i]*(mod_ekn*(mod_ekn+2*m_nucl)) / (ek_galprop[i] * (ek_galprop[i] + 2*m_nucl))
        
            ek_galprop_mod[ia][i] = mod_ekn
            fluxes_galprop_mod[ia][i] = mod_flux
```

```python
istart = 65
fluxes_galprop_mod_tck = [interpolate.UnivariateSpline(np.log(ek_galprop_mod[ia][istart:]), np.log(fluxes_galprop_mod[ia][istart:]), k=3, s=0.001, ext=3) for ia in range(2)]
```
